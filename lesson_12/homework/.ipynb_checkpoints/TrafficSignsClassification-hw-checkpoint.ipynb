````python
import os
import cv2
from time import time
import numpy as np
from sklearn.utils import shuffle

seed = 1 # seed for all random operations

from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = [15, 6]
````

#### Data Loading

Let's now load the data to see what we are dealing with.

````python
def get_images_labels(folder,class_id_folder,class_value):
    # Load traffic sign class
    path_to_class = os.path.join(folder, class_id_folder)
    fnames = os.listdir(path_to_class)
    full_pathes = [ os.path.join(path_to_class,fname) for fname in fnames ]
    only_files = [ full_path  for full_path in full_pathes if os.path.isfile(full_path) ]   # filter directories
    images = [cv2.imread(only_file, cv2.IMREAD_UNCHANGED) for only_file in only_files]
    labels = [class_value] * len(images)
    return images, labels

def print_img_table_random(plt, images, labels, seed):
    np.random.seed(seed)

    for cnt, idx in enumerate(np.random.randint(0, len(images), 10)):
        plt.subplot(2,5,cnt+1)
        plt.imshow(images[idx], cmap='gray', vmin=0, vmax=255)
        plt.title(labels[idx]), plt.axis(False)
        
# functions below just for test if form of flattening influence on result
def spiral_flatten_center(arr):  
    rows, cols = arr.shape
    center_row, center_col = rows // 2, cols // 2
    
    if rows % 2 == 0:
        center_row -= 1
    if cols % 2 == 0:
        center_col -= 1
    
    result = []
    visited = set()
    direction = [(0, 1), (1, 0), (0, -1), (-1, 0)]  # Right, Down, Left, Up
    steps = 1
    row, col = center_row, center_col
    result.append(arr[row, col])
    visited.add((row, col))
    
    while len(result) < rows * cols:
        for d in direction:
            for _ in range(steps):
                row += d[0]
                col += d[1]
                if 0 <= row < rows and 0 <= col < cols and (row, col) not in visited:
                    result.append(arr[row, col])
                    visited.add((row, col))
                if len(result) == rows * cols:
                    break
            if d == direction[1] or d == direction[3]:
                steps += 1

    return np.array(result)

def zigzag_flatten(arr):
    rows, cols = arr.shape
    result = []
    for i in range(rows + cols - 1):
        if i % 2 == 0:
            # Traverse down-right
            x = 0 if i < cols else i - cols + 1
            y = i if i < cols else cols - 1
            while x < rows and y >= 0:
                result.append(arr[x, y])
                x += 1
                y -= 1
        else:
            # Traverse up-left
            x = i if i < rows else rows - 1
            y = 0 if i < rows else i - rows + 1
            while x >= 0 and y < cols:
                result.append(arr[x, y])
                x -= 1
                y += 1
    return np.array(result)

# Example usage
array = np.array([
    [1, 2, 3, 33],
    [4, 5, 6, 66],
    [7, 8, 9, 99]
])

# flattened = spiral_flatten_center(array)
# print(flattened)
# print(spiral_flatten_center(box))
# print(zigzag_flatten(array))


folder = '../data/subset_homework'

images_0, labels_0 = get_images_labels(folder,'class_id_0',0)
print_img_table_random(plt, images_0, labels_0, seed)
````

````python
# Load traffic sign class 1
images_1, labels_1 =  get_images_labels(folder,'class_id_1',1)
print_img_table_random(plt, images_1, labels_1, seed)
````

#### Prepare Input Data

To train our neural network model, we have to prepare the data to the format the the model actually expects. In our case, this will be numpy arrays.

````python
# Put both classes together and shuffle the data
images = images_0 + images_1
labels = labels_0 + labels_1

images, labels = shuffle(images, labels, random_state = seed)

images = np.array(images)
labels = np.array(labels)

print('Images', images.shape)
print('Labels', labels.shape)
````

But now we have a problem. We cannot just feed the image to a neuron since the neuron inputs are flat (one dimensional). On the other hand, the images are 2D matrices. Therefore, we need to "flatten" the images to a one dimensional vector of pixels.

````python
start = time()
pixels = np.array([image.flatten() for image in images])/255
# pixels = np.array([spiral_flatten_center(image) for image in images])/255
stop = time()

print('Shape', pixels.shape)
print('Elapsed time', stop - start)
````

Before the training, let's again have a look at some raqndom samples from our dataset.

````python
print_img_table_random(plt, images, labels, seed)
````

#### Building the Neural Network

### Step 2
Run the training with one single neuron (as we did in the lecture). What accuracy can we achieve?

````python
import tensorflow as tf
from tensorflow.keras import Model
from tensorflow.keras.layers import Dense, Input
````

The following netowork will consist of only one single neuron. It is a very tiny network (not even a network, strictly speaking :-) ) and yet it can be quite powerful.

````python
def build_model(*neuron_array, shape=(10,)):
    inputs = Input(shape=shape)  #shape=(pixels.shape[1],) 
    outputs = inputs
    for neuron_number,activation in neuron_array:
        outputs = Dense(neuron_number, activation=activation)(outputs)
    return Model(inputs, outputs)

def train_model(model, pixels, labels, epochs=10, batch_size=32, verbose=0):
    model.compile(optimizer ='adam', loss = 'mean_squared_error')
    history = model.fit(pixels, labels, epochs=10, batch_size=32,verbose=verbose)
    return model

def get_model_accuracy(model, pixels, labels):
    # Run it on the entire dataset
    predictions = model.predict(pixels).squeeze()
    predictions = predictions > 0.5
    correct = 0
    for prediction, label in zip(predictions, labels):
        if prediction == label:
            correct = correct + 1
    accuracy = correct/len(labels)
    print('Accuracy', accuracy)
    return accuracy

np.random.seed(seed)
tf.random.set_seed(seed)

pixels = np.array([image.flatten() for image in images])/255
pixel_spirals = np.array([spiral_flatten_center(image) for image in images])/255
pixel_zigzags = np.array([zigzag_flatten(image) for image in images])/255

def experiment_accur(pixels):
    model = build_model((1, "linear"),shape=(pixels.shape[1],))
    model = train_model(model, pixels, labels)
    return get_model_accuracy(model, pixels, labels)
max_experements = 10
acc_pixs = [ experiment_accur(pixels) for _ in range(max_experements ) ]
acc_spirals = [ experiment_accur(pixel_spirals) for _ in range(max_experements ) ]
acc_zig = [ experiment_accur(pixel_zigzags) for _ in range(max_experements ) ]

print(f"Acurecies pixels {acc_pixs} np.mean(experiment) = {np.mean(acc_pixs)}" )
print(f"Acurecies pixel spirals {acc_spirals} np.mean(experiment) = {np.mean(acc_spirals)}" )
print(f"Acurecies pixel zigzags {acc_zig} np.mean(experiment) = {np.mean(acc_zig)}" )


# inputs = Input(shape=(pixels.shape[1],))
# outputs = Dense(1, activation="linear")(inputs)
# model = Model(inputs, outputs)

# model.summary()
````

### Conclusion one neuron

As we can see, the average accuracy of one neuron is about 90 percent. Moreover, different forms of representation of the alignment of picture pixels into a one-dimensional array do not have an serious effect.

### Step 3

Make further modifications to improve the accuracy (e.g. add more neurons, more layers, etc.). See what is the maximum accuracy we can achieve?

#### Add linear layers

````python
np.random.seed(seed)
tf.random.set_seed(seed)

pixels = np.array([image.flatten() for image in images])/255

def experiment_accur(pixels, layers):
    model = build_model(*layers,shape=(pixels.shape[1],))
    model = train_model(model, pixels, labels)
    return get_model_accuracy(model, pixels, labels)

layers = (20,"linear"),(1, "linear")
print(f"add layer 20 neurons. Mean accuracy = {np.mean([experiment_accur(pixels, layers) for _ in range(10)])}")

layers = (40,"linear"),(20,"linear"),(1, "linear")
print(f"layers 40,20 neurons. Mean accuraccy = {np.mean([experiment_accur(pixels, layers) for _ in range(10)])}")

layers = (80,"linear"),(40,"linear"),(20,"linear"),(1, "linear")
print(f"layers 80, 40,20 neurons. Mean accuraccy = {np.mean([experiment_accur(pixels, layers) for _ in range(10)])}")
    
````

### Conclusions for adding additional linear layers

Additional layers increase accuracy. But I think that we will win about 5-6 percent with them. 3 layers compared to 2 layers do not give a significant increase.
